```r
data = read.csv('airbnb.csv', header=T)
names(data)
```

```r
data_full = na.omit(data)
```

```r
data_full
```

```r
data = data_full[sample(nrow(data_full), 5000), ]
```

```r
data
```

```r
levels(data$city)
length(levels(data$city))
```

```r
data$city_idx = lapply(data$city, as.numeric)
```

```r
library("rjags")
```

```r
mod_string = " model {
    for (i in 1:length(log_price)) {
        log_price[i] ~ dnorm(mu[city_idx[i]], prec)
        mu[i] = a[city_idx[i]]
                + b[1] * property_type[i]
                + b[2] * room_type[i]
                + b[3] * accommodates[i]
                + b[4] * bathrooms[i]
                + b[5] * bed_type[i]
                + b[6] * cancellation_policy[i]
                + b[7] * bedrooms[i]
                + b[8] * beds[i]
                + b[9] * description_word_count[i]
                + b[10] * description_sent_count[i]
                + b[11] * description_unique_words[i]
                + b[12] * description_sentiment_polarity[i]
                + b[13] * name_word_count[i]
                + b[14] * name_sent_count[i]
                + b[15] * name_unique_words[i]
                + b[16] * name_sentiment_polarity[i]
                + b[17] * review_scores_rating[i]
    }

    for (j in 1:6) {
        a[j] ~ dnorm(a0, prec_a)
    }
    for (j in 1:17) {
        b[j] ~ dnorm(0.0, 1.0/1.0e6)
    }

    a0 ~ dnorm(0.0, 1.0/1.0e6)
    prec_a ~ dgamma(1 / 2.0, 1 * 10.0 / 2.0)
    prec ~ dgamma(5/2.0, 5 *10.0 / 2.0)
    sig = sqrt(1.0 / prec)
    tau = sqrt(1.0 / prec_a)
}
"

set.seed(102)

data_jags = as.list(data)

params = c("a", "a0", "b", "sig", "tau")

mod = jags.model(textConnection(mod_string), data=data_jags, n.chains=3)
update(mod, 1e4)

mod_sim = coda.samples(model=mod,
                        variable.names=params,
                        n.iter=5e3)
mod_csim = as.mcmc(do.call(rbind, mod_sim))

## convergence diagnostics
plot(mod_sim)

gelman.diag(mod_sim)
autocorr.diag(mod_sim)
autocorr.plot(mod_sim)
effectiveSize(mod_sim)

## compute DIC
dic = dic.samples(mod, n.iter=1e3)
dic
```
