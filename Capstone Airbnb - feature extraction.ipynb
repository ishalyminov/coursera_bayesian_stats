```python
import numpy as np
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import sent_tokenize

nltk.download('vader_lexicon')
```

```python
REVIEWS_FILE = 'reviews.csv'
data = pd.read_csv(REVIEWS_FILE)
```

```python
def annotate_sentiment(in_list):
    sent = SentimentIntensityAnalyzer()
    sentiment = [sent.polarity_scores(item) for item in in_list]
    sentiment_polarity = [item['compound'] for item in sentiment]
    return sentiment_polarity
```

```python
def annotate_textual_features(in_list):
    result = {'word_count': [], 'sent_count': [], 'unique_words': [], 'sentiment_polarity': []}

    tokenizer = TweetTokenizer()
    for item_idx, item in enumerate(in_list):
        # print('Processing item {} of {}'.format(item_idx, len(in_list)))
        words = tokenizer.tokenize(item)

        result['word_count'].append(len(words))
        result['sent_count'].append(len(sent_tokenize(item)))
        result['unique_words'].append(len(set(words)))
    result['sentiment_polarity'] = annotate_sentiment(in_list)

    return result
```

```python
textual_features = annotate_textual_features(data['description'])
for feat, value in textual_features.items():
    data['description_' + feat] = value
```

```python
textual_features = annotate_textual_features(data['name'])
for feat, value in textual_features.items():
    data['name_' + feat] = value
```

```python
data = data.drop('description', axis=1).drop('name', axis=1)
```

```python
data.to_csv('airbnb.csv')
```
